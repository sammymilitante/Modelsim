```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
# To read image from disk, we use
# cv2.imread function, in below method,
img = cv2.imread("Computer_vision.jpg", cv2.IMREAD_COLOR)
plt.title("Computer Vision")
plt.imshow(img)
plt.show()
```
