#Image Thresholding

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img_th = cv2.imread('cat.jpg',0)
```

```python
ret,thresh1 = cv2.threshold(img_th,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img_th,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img_th,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img_th,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img_th,127,255,cv2.THRESH_TOZERO_INV)
```

```python
plt.title("Original Image")
plt.imshow(img_th)
```

```python
plt.title("Binary Image")
plt.imshow(thresh1)
```

```python
plt.title("Binary_inv Image")
plt.imshow(thresh2)
```

```python
plt.title("Trunc Image")
plt.imshow(thresh3)
```

```python
plt.title("To_zero Image")
plt.imshow(thresh4)
```

```python
plt.title("To_zero_inv Image")
plt.imshow(thresh5)
```
