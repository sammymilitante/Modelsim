# Morphological Transformation

```python

import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat.jpg',0)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(thresh2,kernel,iterations = 2)
dilation = cv2.dilate(thresh2,kernel,iterations = 3)
```

```python
plt.title("Original Image")
plt.imshow(img)
```

```python
plt.title("Thresh2")
plt.imshow(thresh2)
```

```python
plt.title("Erosion")
plt.imshow(erosion)
```

```python
plt.title("Dilation")
plt.imshow(dilation)
```
